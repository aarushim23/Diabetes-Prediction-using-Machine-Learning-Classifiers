```python
import pandas as pd
from factor_analyzer import FactorAnalyzer
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the diabetes dataset
data =pd.read_csv(r'C:\Users\aarus\Downloads\diabetes-dataset.csv')
data.head()
```

```python
# Select the relevant columns (e.g., 'Pregnancies', 'Age', and 'Outcome')
selected_columns = ['Pregnancies', 'Age', 'Outcome']
data = data[selected_columns].copy()
```

```python
# Drop rows with NaN values
data = data.dropna()
```

```python
# Perform factor analysis
fa = FactorAnalyzer(n_factors=1, rotation=None)
fa.fit(data)
```

```python
# Get the factor loadings
loadings = fa.loadings_

# Print the factor loadings
print("Factor Loadings:")
print(loadings)
```

```python
# Normalize the 'Outcome' column for assigning colors
normalized_outcome = (data['Outcome'] - data['Outcome'].min()) / (data['Outcome'].max() - data['Outcome'].min())
```

```python
# Plot the scatter plot with color gradients based on 'Outcome'
plt.scatter(data['Pregnancies'], data['Age'], c=normalized_outcome, cmap='coolwarm')
plt.xlabel('Number of Pregnancies')
plt.ylabel('Age')
plt.title('Diabetes Risk: Pregnancies vs Age')
plt.colorbar(label='Normalized Outcome')
plt.show()
```

```python
import pandas as pd
from scipy.stats import pearsonr
from scipy.stats import ttest_ind

# Load the diabetes dataset
data =pd.read_csv(r'C:\Users\aarus\Downloads\diabetes-dataset.csv')

# Select the relevant columns (e.g., 'Pregnancies', 'Age', and 'Outcome')
selected_columns = ['Pregnancies', 'Age', 'Outcome']
data = data[selected_columns].copy()

# Drop rows with missing values
data = data.dropna()

# Calculate the correlation coefficient
corr_coefficient, p_value = pearsonr(data['Pregnancies'], data['Age'])
print(f"Correlation Coefficient: {corr_coefficient:.3f}")
print(f"P-value: {p_value:.3f}")

# Perform independent t-test
diabetic_pregnancies = data[data['Outcome'] == 1]['Pregnancies']
non_diabetic_pregnancies = data[data['Outcome'] == 0]['Pregnancies']
t_statistic, p_value = ttest_ind(diabetic_pregnancies, non_diabetic_pregnancies)
print(f"T-statistic: {t_statistic:.3f}")
print(f"P-value: {p_value:.3f}")
```
